```python
train.dtypes
```

```python
all_data = pd.concat([train,test])
```

```python
all_data2 = all_data[all_data.columns[all_data.dtypes != 'object']]
all_data2
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1)
rf.fit(train2, train['interest_level'])
result = rf.predict_proba(test2)
result
```

```python
sub[['high', 'medium', 'low']] = result ## iloc으로 접근해서 넣어도 됨. [:,1:] = result
sub
```

```python
sub.to_csv('sub.csv', index = 0)
```

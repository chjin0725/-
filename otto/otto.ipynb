```python
all_data2 = all_data.drop(['id', 'target'], axis=1)
```

```python
# from sklearn.preprocessing import StandardScaler
# ss = StandardScaler()
# from sklearn.preprocessing import RobustScaler ## ê°’ì´ ì–‘ìª½ìœ¼ë¡œ íŠ€ëŠ”ã„± ã…”ì•„ë‹ˆë¼ í•œìª½ì— íŠ„ë‹¤ --> robust scalerë³´ë‹¤ëŠ” logë¥¼ ì”Œì›Œì¤€ë‹¤.
# rs = RobustScaler()
# all_data3 = rs.fit_transform(all_data2)
# pd.DataFrame(all_data3)
all_data3 = np.log1p(all_data2) ## log1p = log(x+1)ì„ ì˜ë¯¸.
all_data3
```

```python
train2 = all_data3[:len(train)]
test2 = all_data3[len(train):]
```

```python
# from keras import Sequential
# from keras.layers import *
# model = Sequential()
# model.add(Dense(1536, activation = 'relu', input_dim = train2.shape[1]))# ì˜µì…˜ ìµœì†Œ 2ê°€ì§€ ë“¤ì–´ê°. ì•„ì›ƒí’‹ í¬ê¸°ë¥¼ 1024ì •ë„ë©´ ì›¬ë§Œí•˜ë©´ ì˜ ë‚˜ì˜¨ë‹¤. input layerì´ë¯€ë¡œ inputì˜ ì°¨ì›ë„ ë„£ê¸°.
# model.add(Dropout(0.3))
# model.add(Dense(768, activation = 'relu'))
# model.add(Dropout(0.3))
# model.add(Dense(256, activation='relu'))
# model.add(Dropout(0.3))
# model.add(Dense(9, activation = 'softmax'))## ì´ì§„ë¶„ë¥˜ë„ ì†Œí”„íŠ¸ë§¥ìŠ¤ë¡œ 

# model.compile(metrics=['acc'], optimizer='adam', loss='sparse_categorical_crossentropy') ## sparse_categorical_crossentropyë¡œ (sparse ì¶”ê°€) ì ì–´ì•¼ ë¶„ë¥˜ë¡œ í•™ìŠµí•œë‹¤.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(train['target']) ## ìŠ¤ì¼€ì¼ë§ í›„ì˜ yëŠ” ë°ì´í„°í”„ë ˆì„ì´ ì•„ë‹ˆë¼ ê·¸ëƒ¥ ë°°ì—´ì„.
y
# from sklearn.model_selection import train_test_split
# x_train, x_valid, y_train, y_valid = train_test_split(train2, y, test_size=0.2, random_state=42, stratify=y) ## stratifyë¥¼ ì¨ì•¼ train2ì™€ yì—ì„œ ê° í´ë˜ìŠ¤ì˜ ë¹„ìœ¨ì´ ê°™ê²Œ ëœë‹¤. train,validì— ê°™ì€ ë¹„ìœ¨ë¡œ í´ë™ìŠ¤ë¥¼ ë‚˜ëˆ”
```

```python
from sklearn.model_selection import StratifiedKFold ## íšŒê·€ì—ì„œë„ ì´ëŸ°ì‹ìœ¼ë¡œ í•  ìˆ˜ ìˆë‹¤. yê°’ì˜ ë²”ìœ„ê°€ 30~40ì— ë§ìœ¼ë©´ ì´ìª½ ë²”ìœ„ì˜ yë¥¼ ë¹„ìœ¨ì— ë§ê²Œ ë” ë½‘ì„ ìˆ˜ ìˆìŒ.
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

from keras import Sequential
from keras.layers import *

final_result = 0
for train_index, valid_index in skf.split(train2, y): # 80%ëŠ” trainì— 20%ëŠ” valid
    print(valid_index)
    x_train = train2.iloc[train_index]
    x_valid = train2.iloc[valid_index]
    y_train = y[train_index] ## ë°ì´í„°í”„ë ˆì„ì´ ì•„ë‹ˆë¯€ë¡œ ê·¸ëƒ¥ ëŒ€ê´„í˜¸ë¥¼ ì“´ë‹¤.
    y_valid = y[valid_index]
    
    model = Sequential()
    model.add(Dense(1536, activation = 'relu', input_dim = train2.shape[1]))# ì˜µì…˜ ìµœì†Œ 2ê°€ì§€ ë“¤ì–´ê°. ì•„ì›ƒí’‹ í¬ê¸°ë¥¼ 1024ì •ë„ë©´ ì›¬ë§Œí•˜ë©´ ì˜ ë‚˜ì˜¨ë‹¤. input layerì´ë¯€ë¡œ inputì˜ ì°¨ì›ë„ ë„£ê¸°.
    model.add(Dropout(0.3))
    model.add(Dense(768, activation = 'relu'))
    model.add(Dropout(0.3))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(9, activation = 'softmax'))## ì´ì§„ë¶„ë¥˜ë„ ì†Œí”„íŠ¸ë§¥ìŠ¤ë¡œ. softmaxì—ì„œ í´ë˜ìŠ¤ ê°œìˆ˜ê°€ 2ê°œì´ë©´ sigmoidê°€ ë‚˜ì˜¨ë‹¤. softmaxëŠ” ë¶„ë¥˜ì— ì“°ì´ëŠ” í•¨ìˆ˜ì˜ ì¼ë°˜í˜• ê°™ì€ê±°ì„. 

    model.compile(metrics=['acc'], optimizer='adam', loss='sparse_categorical_crossentropy')
    
    model.fit(x_train, y_train, validation_data = (x_valid, y_valid), epochs=6, batch_size = 128) ## validation_splitì€ ë§¨ ë’¤ì—ì‡ëŠ” ê²ƒ 0.1ì„ validë¡œ ì“´ë‹¤. ì¦‰ ì…”í”Œì„ ì•ˆí•¨. ê·¸ë˜ì„œ ê·¸ëƒ¥ í•˜ë©´ val_lossê°’ì´ ë§¤ìš° í¬ë‹¤
    
    final_result += model.predict(test2)/5 ## final_resultëŠ” ì–´ë ˆì´ í˜•íƒœì„. ëª¨ë“  í…ŒìŠ¤íŠ¸ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡ê°’. ì—¬ê¸°ì„  ê·¸ëƒ¥ 5ë¡œ ë‚˜ëˆ„ê³  ìˆëŠ”ë° ì›¨ì´íŠ¸ë¥¼ ì¤˜ì„œ ì•™ìƒë¸”ì„ í•  ìˆ˜ë„ ìˆìŒ.


```

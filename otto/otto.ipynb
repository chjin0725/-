```python
all_data2 = all_data.drop(['id', 'target'], axis=1)
```

```python
# from sklearn.preprocessing import StandardScaler
# ss = StandardScaler()
# from sklearn.preprocessing import RobustScaler ## 값이 양쪽으로 튀는ㄱ ㅔ아니라 한쪽에 튄다 --> robust scaler보다는 log를 씌워준다.
# rs = RobustScaler()
# all_data3 = rs.fit_transform(all_data2)
# pd.DataFrame(all_data3)
all_data3 = np.log1p(all_data2) ## log1p = log(x+1)을 의미.
all_data3
```

```python
train2 = all_data3[:len(train)]
test2 = all_data3[len(train):]
```

```python
train['target'].value_counts()
```

```python
# from keras import Sequential
# from keras.layers import *
# model = Sequential()
# model.add(Dense(1536, activation = 'relu', input_dim = train2.shape[1]))# 옵션 최소 2가지 들어감. 아웃풋 크기를 1024정도면 웬만하면 잘 나온다. input layer이므로 input의 차원도 넣기.
# model.add(Dropout(0.3))
# model.add(Dense(768, activation = 'relu'))
# model.add(Dropout(0.3))
# model.add(Dense(256, activation='relu'))
# model.add(Dropout(0.3))
# model.add(Dense(9, activation = 'softmax'))## 이진분류도 소프트맥스로 

# model.compile(metrics=['acc'], optimizer='adam', loss='sparse_categorical_crossentropy') ## sparse_categorical_crossentropy로 (sparse 추가) 적어야 분류로 학습한다.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(train['target']) ## 스케일링 후의 y는 데이터프레임이 아니라 그냥 배열임.
y
# from sklearn.model_selection import train_test_split
# x_train, x_valid, y_train, y_valid = train_test_split(train2, y, test_size=0.2, random_state=42, stratify=y) ## stratify를 써야 train2와 y에서 각 클래스의 비율이 같게 된다. train,valid에 같은 비율로 클랙스를 나눔
```

```python
from sklearn.model_selection import StratifiedKFold ## 회귀에서도 이런식으로 할 수 있다. y값의 범위가 30~40에 많으면 이쪽 범위의 y를 비율에 맞게 더 뽑을 수 있음.
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

from keras import Sequential
from keras.layers import *

final_result = 0
for train_index, valid_index in skf.split(train2, y): # 80%는 train에 20%는 valid
    print(valid_index)
    x_train = train2.iloc[train_index]
    x_valid = train2.iloc[valid_index]
    y_train = y[train_index] ## 데이터프레임이 아니므로 그냥 대괄호를 쓴다.
    y_valid = y[valid_index]
    
    model = Sequential()
    model.add(Dense(1536, activation = 'relu', input_dim = train2.shape[1]))# 옵션 최소 2가지 들어감. 아웃풋 크기를 1024정도면 웬만하면 잘 나온다. input layer이므로 input의 차원도 넣기.
    model.add(Dropout(0.3))
    model.add(Dense(768, activation = 'relu'))
    model.add(Dropout(0.3))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(9, activation = 'softmax'))## 이진분류도 소프트맥스로. softmax에서 클래스 개수가 2개이면 sigmoid가 나온다. softmax는 분류에 쓰이는 함수의 일반형 같은거임. 

    model.compile(metrics=['acc'], optimizer='adam', loss='sparse_categorical_crossentropy')
    
    model.fit(x_train, y_train, validation_data = (x_valid, y_valid), epochs=6, batch_size = 128) ## validation_split은 맨 뒤에잇는 것 0.1을 valid로 쓴다. 즉 셔플을 안함. 그래서 그냥 하면 val_loss값이 매우 크다
    
    final_result += model.predict(test2)/5 ## final_result는 어레이 형태임. 모든 테스트데이터에 대한 예측값. 여기선 그냥 5로 나누고 있는데 웨이트를 줘서 앙상블을 할 수도 있음.


```
